```python
# 첫번째 방법
import csv
f = open(r'../데이터분석/gender.csv', encoding='cp949')
data = csv.reader(f)
m = []
f = []

for row in data:
    if '와부읍' in row[0]:
        for i in range(0,101):
            m.append(int(row[i+3]))
            f.append(int(row[-(i+1)]))
f.reverse() # 여자 역순으로 재배열
```

```python
# 두번째 방법
import csv
f = open(r'../데이터분석/gender.csv', encoding='cp949')
data = csv.reader(f)
m = []
f = []

for row in data:
    if '와부읍' in row[0]:
        for i in row[3:104]:
            m.append(-int(i))
        for i in row[106:]:
            f.append(int(i))
```

```python
import matplotlib.pyplot as plt
plt.barh(range(101), m, color='b', label="남")
plt.barh(range(101), f, color='r', label="여")
plt.legend()
plt.rc('font',family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False # 마이너스 표시
plt.show()
```

### 우리 동네 인구구조를 항아리 모양 그래프로 시각화

```python
import csv
import matplotlib.pyplot as plt

f = open('../데이터분석/gender.csv', encoding='cp949')
data = csv.reader(f)

m = []
f = []
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')
for row in data:
    if name in row[0]:
        for i in row[3:104]:
            m.append(-int(i))
        for i in row[106:]:
            f.append(int(i))
        break   
plt.style.use('ggplot')
plt.figure(figsize=(10,5))
plt.rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name + ' 지역의 남녀 성별 인구 분포')
plt.barh(range(101), m, label='남성')
plt.barh(range(101), f, label="여성")
plt.xlabel('인구수')
plt.ylabel('연령대')
plt.legend()
plt.show()
```

### pie chart

```python
import csv
import matplotlib.pyplot as plt

f = open('../데이터분석/gender.csv', encoding='cp949')
data = csv.reader(f)
size = []
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')
for row in data:
    if name in row[0]:
        m = 0
        f = 0
        for i in range(101):
            m += int(row[i+3]) # 3~103번까지 데이터
            f += int(row[i+106])
        break
size.append(m)
size.append(f)
plt.rc('font',family='AppleGothic')
color = ['skyblue','pink']
plt.axis('equal')
plt.pie(size, labels=['남','여'], autopct='%.1f%%', colors=color, startangle=90) # 파이차트의 시작각도 90도
plt.title(name + ' 지역의 남녀 성별 비율')
plt.show()
```
