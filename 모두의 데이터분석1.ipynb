```python
import csv
import matplotlib.pyplot as plt

f = open('/seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #header  
high = []
low = []

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '09' and row[0].split('-')[2] == '06': # -    
            high.append(float(row[-1])) #   high  
            low.append(float(row[-2])) #   low  

plt.figure(figsize=(10,3))
plt.plot(high, color='red', label='high', marker='o')
plt.plot(low, color='blue', label='low', marker='v')
plt.xlabel('')
plt.ylabel('')
plt.title('    ')
plt.legend()
plt.rc('font',family='AppleGothic')
plt.show()
```
