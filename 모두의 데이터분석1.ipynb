### 내 생일의 기온 변화 그래프

```python
import csv
import matplotlib.pyplot as plt

f = open('데이터분석/seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #header 다음문장부터 표시
high = []
low = []

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '09' and row[0].split('-')[2] == '06': # - 문자를 기준으로 구분하여 저장
            high.append(float(row[-1])) # 최고 기온값을 high 리스트에 저장
            low.append(float(row[-2])) # 최저 기온값을 low 리스트에 저장

plt.figure(figsize=(10,3))
plt.plot(high, color='red', label='high', marker='o')
plt.plot(low, color='blue', label='low', marker='v')
plt.xlabel('나이')
plt.ylabel('온도')
plt.title('내 생일의 기온 변화 그래프')
plt.legend()
plt.rc('font',family='AppleGothic')
plt.show()
```
