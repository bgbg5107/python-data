### 기온 데이터를 다양하게 시각화

```python
import csv
import matplotlib.pyplot as plt

f = open('데이터분석/seoul.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
result = []

for row in data:
    if row[-1] != '':
        result.append(float(row[-1]))
        
plt.hist(result, bins=100) # x축의 구간갯수
plt.show()
```

### 1월과 8월의 데이터를 히스토그램으로 시각화

```python
import csv
import matplotlib.pyplot as plt

f = open('데이터분석/seoul.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
aug = []
jan = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '08':
            aug.append(float(row[-1]))
        if month == '01':
            jan.append(float(row[-1]))
        
plt.hist(aug, bins=100, color='r', label='Aug') # x축의 구간갯수
plt.hist(jan, bins=100, color='b', label='Jan')
plt.legend()
plt.show()
```

### 8월 일별 기온 데이터를 상자그림으로 시각화

```python
import csv
import matplotlib.pyplot as plt

f = open('데이터분석/seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)

day = []
for i in range(31): # 일별 데이터를 저장할 리스트 day 생성 후 반복문으로 31개 리스트 생성
    day.append([])
    
for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08':
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))
            
plt.style.use('ggplot') # 그래프 스타일 지정
plt.figure(figsize=(10,5))
plt.xlabel('8월')
plt.boxplot(day, showfliers=False) # 이상치 제거
plt.rc('font',family='AppleGothic')
plt.show()
```
